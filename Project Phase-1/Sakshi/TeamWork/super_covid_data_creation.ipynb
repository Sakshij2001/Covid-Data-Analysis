## Covid Deaths DataSet Display

## Merging All the 3 Datasets Based On County-FIPS and State  To Create A Dataframe

```python
#Merging all the covid deaths and covid confirmed data sets based on countyFIPS and State.
#Naming the columns of the covid deaths dataset with _deaths suffix and covid confirmed data sets with _cases suffix
merged_df = data_covid_deaths.merge(data_covid_cases, on = ("countyFIPS", "State"), suffixes = ("_deaths" , "_cases"))
```

```python
#Merging all above merged_df dataset with covid county population dataset based on countyFIPS and State.
#Naming the columns of the covid county population dataset with the suffix _population
merged_df = merged_df.merge(data_county_population, on = ("countyFIPS", "State"), suffixes = ("","_population"))
```
